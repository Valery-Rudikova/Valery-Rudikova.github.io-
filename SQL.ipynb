С целью минимизации кода создадим функцию, которая будет возвращать объект DataFrame, принимая строку с SQL запросом.

```python
def request(str):
    return pd.read_sql('''{}'''.format(str), connect)
```

```python
request('''
SELECT BillingCountry, BillingCity, SUM(total), COUNT(total), STRFTIME("%Y", invoiceDate) AS year
FROM invoices
GROUP BY BillingCountry, BillingCity
HAVING SUM(total) > 40 AND year != '2009'
ORDER BY COUNT(total) DESC
''')
```

Задание.<br>
Найдите топ-5 клиентов(фамилия и имя) по среднему чеку заказа. Для этого присоедините таблицу customers к таблице invoice.

```python
request('''
SELECT FirstName, LastName, AVG(UnitPrice)
FROM invoices AS i
INNER JOIN invoice_items AS ii ON i.InvoiceId = ii.InvoiceId
INNER JOIN customers AS c ON i.CustomerId = c.CustomerId
GROUP BY c.FirstName, c.LastName
ORDER BY AVG(UnitPrice) DESC
LIMIT 5
''')
```

Задание:

```python
request('''
SELECT ar.Name, COUNT(DISTINCT Title)
FROM tracks AS t 
INNER JOIN albums AS a ON t.albumId = a.albumId
INNER JOIN artists AS ar ON a.artistId = ar.artistId
GROUP BY ar.Name
ORDER BY COUNT(DISTINCT Title) DESC
LIMIT 5
''')
```
